## 打标签

```python
import numpy as np
def label_build(begin,end):
    label_y = np.zeros(end-begin+980,dtype='float')
    for i in range(end-begin+980):
        if i < 200:
            label_y[i] = 780
        elif i < 980 and i>= 200:
            label_y[i] = 980-i
        else:
            label_y[i] = -1
    label_y[0:980] = label_y[0:980]/780
    label_y = pd.DataFrame(label_y)
    return label_y
```

```python
train_y1 = label_build(85093,85111)
train_y1.columns = error
train_y2 = label_build(1061474,1061494)
train_y2.columns = error
test_y = label_build(1282046,1282088)
test_y.columns = error
```

```python
train_y = pd.concat([train_y1,train_y2],axis=0)
train_y = train_y.reset_index(drop=True)
```

```python
train_y.plot()
```

```python
train_x1 = df[feature_list][85093-980:85111]
train_x2 = df[feature_list][1061474-980:1061494]
test_x = df[feature_list][1282046-980:1282088]
```

```python
train_x = pd.concat([train_x1, train_x2], axis=0)
train_x = train_x.reset_index(drop=True)
```

## 生成sequence

```python
sequence_length = 50
```

```python
# function to reshape features into (samples, time steps, features)
def gen_sequence(id_df, seq_length, seq_cols):
    """ Only sequences that meet the window-length are considered, no padding is used. This means for testing
    we need to drop those which are below the window-length. An alternative would be to pad sequences so that
    we can use shorter ones """
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        yield data_array[start:stop, :]
```

```python
# generator for the sequences
seq_gen = list(gen_sequence(train_x, 999, feature_list))
           
```

```python
# generate sequences and convert to numpy array
seq_array = np.concatenate(list(seq_gen)).astype(np.float32)
seq_array.shape
#train_x1.shape
#seq = (list(train_x1),list(train_x2))
#train_x = np.concatenate(list(seq)).astype(np.float32)
```

```python
test_seq_gen = tuple([list(gen_sequence(test_x, sequence_length, feature_list))])
test_seq_array = np.concatenate(list(test_seq_gen)).astype(np.float32)
test_seq_array.shape
```

```python
# function to generate labels
def gen_labels(id_df, seq_length, label):
    data_array = id_df[label].values
    num_elements = data_array.shape[0]
    return data_array[seq_length:num_elements, :]
```

```python
# generate labels
label_gen = [gen_labels(train_y1, sequence_length, error),
             gen_labels(train_y2, sequence_length, error)]
label_array = np.concatenate(label_gen).astype(np.float32)
label_array.shape
```

```python
test_label_gen = [gen_labels(test_y, sequence_length, error)]
test_label_array = np.concatenate(test_label_gen).astype(np.float32)
test_label_array.shape
```

## 训练

```python
import keras
```

```python
from keras.layers import Dense, Dropout, LSTM, Activation
from keras.models import Sequential
from sklearn.metrics import confusion_matrix, recall_score, precision_score
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Setting seed for reproducability
np.random.seed(1234)
PYTHONHASHSEED = 0
```

```python
nb_features = seq_array.shape[2]
nb_out = label_array.shape[1]

model = Sequential()

model.add(LSTM(
    input_shape=(999, 23),
    units=100,
    return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
          units=999,
          return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=nb_out, activation='softmax'))
model.compile(loss='mse',
              optimizer='adam', metrics=['accuracy'])
nb_features
nb_out
```

```python
% %time
# fit the network
model.fit(seq_array, label_array, epochs=10, batch_size=200, validation_split=0.05, verbose=1)
```

```python
# training metrics
scores = model.evaluate(seq_array, label_array, verbose=1, batch_size=200)
print('Accurracy: {}'.format(scores[1]))
```

## 测试
